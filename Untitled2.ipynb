```python
import tensorflow as tf
import numpy as np

# Create a simple model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(5,)),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Save the model
model.save("model.h5")
print("Model has been saved as 'model.h5'")
```

```python
import tensorflow as tf
from flask import Flask, request, jsonify

# Custom objects (in this case, the 'mse' metric)
custom_objects = {'mse': tf.keras.losses.MeanSquaredError()}

# Load your trained model (ensure your .h5 file name matches)
model = tf.keras.models.load_model("model.h5", custom_objects=custom_objects)

# Initialize Flask app
app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
    # Get JSON data from the request
    input_data = request.get_json()

    # Convert JSON to a format compatible with the model
    # Ensure the input data matches the input shape of your model
    data = tf.constant(input_data)

    # Make a prediction
    prediction = model.predict(data)

    # Return the prediction as JSON
    return jsonify({'prediction': prediction.tolist()})

if __name__ == "__main__":
    app.run(debug=True)
```

```python
import tensorflow as tf
from flask import Flask, request, jsonify, render_template

# Custom objects (in this case, the 'mse' metric)
custom_objects = {'mse': tf.keras.losses.MeanSquaredError()}

# Load your trained model (ensure your .h5 file name matches)
model = tf.keras.models.load_model("model.h5", custom_objects=custom_objects)

# Initialize Flask app
app = Flask(__name__)

@app.route('/')
def home():
    return render_template('index.html')  # Serve the frontend HTML

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get user input from form
        user_input = request.form['input_data']

        # Convert input to a list of floats
        input_data = [[float(x) for x in user_input.split(',')]]

        # Convert data to tensor
        data = tf.constant(input_data)

        # Make a prediction
        prediction = model.predict(data)

        # Return prediction as a string
        return render_template('index.html', prediction_text=f'Prediction: {prediction.tolist()}')

    except Exception as e:
        return render_template('index.html', prediction_text=f'Error: {str(e)}')

if __name__ == "__main__":
    app.run(debug=True)
```
